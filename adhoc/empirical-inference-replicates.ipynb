```python
import pandas as pd
from pathlib import Path
import re

results = list(Path('../output/inferences-empirical-replicates/').glob('log-sel-strength_main-partialsweeps_*'))

df = pd.concat([
    pd.read_table(res).assign(replicate=int(res.stem.split('-')[-1]))
    for res in results
]).reset_index()
df['locus'] = [re.search(r"sweep-(.+?)_", i).groups()[0] for i in df.uuid]
df = df.sort_values(by='locus')
df['pred_s'] = 10**(df.predicted_log_selection_coefficient)
df = df[['locus', 'pred_s', 'replicate']]
df = df.pivot(index='replicate', columns='locus', values='pred_s')

df.describe()
```
